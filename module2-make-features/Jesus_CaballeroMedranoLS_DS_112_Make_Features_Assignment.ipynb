```python
#We need now to write down a fuction that does what have been done.

def date_to_integer(cell):
  return int(cell.strip('months').strip())


```

```python
# Function to define if credit status.


def credit_status(cell):
  if cell == 'Current':
    return 1
  elif cell == 'Paid':
    return 1
  else:
    return 0



```

```python
# Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.

#First define a function that converts the null values to zeros.

def convtozero(cell):
  if pd.notnull(cell):
    return cell
  else:
    return 0

#Then apply this function to the column

df_nan = df['last_pymnt_d'].apply(convtozero)

#Extract the month and the year separately.

pd.to_datetime(df_nan,infer_datetime_format=True).dt.year


pd.to_datetime(df_nan,infer_datetime_format=True).dt.month

#Join them to the data frame.

df['last_pymnt_d_month']= pd.to_datetime(df_nan,infer_datetime_format=True).dt.month

df['last_pymnt_d_year'] = pd.to_datetime(df_nan,infer_datetime_format=True).dt.year


df.head()



```
